## **2. Selection Structures**

```python
# if selestion structure
mark=49
if mark>=50:
    print("You passed this course")
print("End of if structure")
```

```python
# if-else selestion structure
mark=50
if mark>=50:
    print("You passed this course")
else:
    print("You failed this course")
print("End of if structure")
```

```python
# Nested if else structures
midterm=80
final=50
if final>=50:
    print("You passed this course")
elif midterm>=80:
    print("You passed this course")
else:
    print("You failed this course")
print("End of if structure")
```

## **3. Repetition Structures (Loops)**

**3.1. For Loops**

```python
# for loops - 1
for i in range(10):
    print(i)
```

```python
# for loops - 2
for i in range(1,5):
    print(i)
else:
    print("The loop has ended")
```

```python
# for loops - 3
for x in range(6):
  if x == 3: break
  print(x)
else: #The else block will NOT be executed if the loop is stopped by a break statement
  print("Finally finished!")
```

```python
# for loops - 4
for x in range(6):
  if x == 3: continue
  print(x)
else: #The else block will NOT be executed if the loop is stopped by a break statement
  print("Finally finished!")
```

**3.1.1. Example: Fizz-Buzz counting problem**

Counting from 1 to 50, but when a number is a multiple of 3, Fizz is written and when a number is
a multiple of 5, Buzz is written. If a number is multiple of both 3 and 5, then FizBuzz is written.

```python
# Fizz-Buzz counting problem
for i in range(1,21):
    if i%3==0 and i%5==0:
        print("FizzBuzz")
    elif i%3==0:
        print("Fizz")
    elif i%5==0:
        print("Buzz")
    print(i)
```

## **4. Lists in Python**

**4.1. Creating and printing Lists**

```python
# Creating and printing Lists
mylist=["apple", "banana", "cherry", "orange", "pear", "mellon", "kiwi", "mango"]
print(mylist)
```

**4.2. Accessing List Items**

```python
# Accessing list items
print(mylist[0]) # The first item has index 0
print(mylist[7]) # The last item has index 7
print(mylist[2:5]) #The print will start at index 2 (included) and end at index 5 (not included)
print(mylist[:4]) # By leaving out the start value, the range will start at the first item
print(mylist[2:]) # By leaving out the end value, the range will go on to the end of the list
print(mylist[-2]) # Negative indexing means start from the end; -1 refers to the last item, -2 refers to the second last item etc.
```

**4.3 Modifying List Items**

```python
# Modifying list items - 1
mylist[5]= "watermellon" #To change the value of a specific item, refer to the index number
print(mylist)
mylist[2:3]=["blackcurrant","blueberry"] #Change the third value by replacing it with two new values
print(mylist)
mylist[2:4]=["cherry"] #Change the third and fourth value by replacing it with one value
print(mylist)
```

```python
# Modifying list items - 2
mylist =[10, 15, 20, 25, 30, 35, 40]
for x in mylist:
    print(x * 2)
```

**4.4. Slicing Operations with Lists**

```python
list1=[20, "SNET", 8.65, -87, 45, False, "Eureka", 78, True, -12.3, "CEN376"]
print("list1=", list1)

list2=list1[2:5]
print("list2=", list2)

list3=list1[4:]
print("list3=", list3)

list4=list1[ :5]
print("list4=", list4)

list5=list1[2:10:3]
print("list5=", list5)
```

**4.5. Adding new Items to the Lists**

```python
mylist=["apple", "banana", "cherry"]
print(mylist)

mylist.append("orange") #To add an item to the end of the list, use the append() method
print(mylist)

mylist.insert(1,"mellon") #To insert a list item at a specified index, use the insert() method
print(mylist)

tropical=["mango", "pineapple", "papaya"]
mylist.extend(tropical) #To append elements from another list to the current list, use the extend() method
print(mylist,"\n")
```

**4.6. Removing Existing Items from the Lists**

```python
mylist.remove("mellon") # The remove() method removes the specified item. If there are more than one
                        # item with the specified value, the remove() method removes the first occurance
print(mylist)

mylist.pop(5) #The pop() method removes the item with specified index
print(mylist)

mylist.pop() #If you do not specify the index, the pop() method removes the last item
print(mylist)
```

**4.7. Clearing or Deleting Entire List**

```python
del mylist[3]  #The del keyword also removes the specified index
print(mylist)

del tropical  #The del keyword can also delete the list completely

mylist.clear() #The clear() method empties the list
print(mylist)
```
