## **1.1. Creating a DataFrame from a list of lists**

One way to create a DataFrame is by using a list of lists. Each inner list represents a row of data, and the outer list encompasses all the rows. You can create a DataFrame by simply passing this list to pd.DataFrame().

```python
import pandas as pd
my_list = [ [1,2,3], [9,10,12], [20,30,40], [60, 75, 85] ]
df = pd.DataFrame(my_list)
df
```

To specify column and row names:
- The columns parameter allows you to set names for the columns, while the index sets custom row labels.

```python
# Specifying column names
df2 = pd.DataFrame(my_list, columns=['A','B','C'])
df2
```

```python
# Specifying column names and row names (indexes)
df3 = pd.DataFrame(my_list, columns=['A','B','C'], index = ['JAN', 'FEB', 'MAR','APR'])
df3
```

## **1.2. Create a DataFrame using dictionary**

- In this method, a dictionary is used to define the columns and their corresponding values (lists).
- Here, the dictionary keys become the column labels, and the values (lists) become the rows for each column.
- The r list provides custom row labels (JAN, FEB, MAR, etc.).

```python
d = {'column1': [1, 2, 3, 4, 7], 'column2': [4, 5, 6, 9, 5], 'column3': [7, 8, 12, 1, 11]}
r = ['JAN', 'FEB', 'MAR', 'APR', 'MAY']

df = pd.DataFrame(data=d, index=r)
df
```

- **iloc[ ]** selects rows based on integer positions (index number).
- **loc[ ]** selects rows based on label (index label).

Both will return the same result here since the index is labeled numerically.

```python
# First row by index position
mydatairis.iloc[0]
```

```python
# First row by label
mydatairis.loc[0]
```

***Why is this a DataFrame?***
- By passing the column index as a list ([0]), you're telling Pandas to treat the selection as a DataFrame (even if it has only one column).

```python
#A list of indexes [[0, 2]] returns a Pandas DataFrame.
# First column
print(type(mydatairis.iloc[: ,[0]]))
print("Shape: ",mydatairis.iloc[:, [0]].shape)
mydatairis.iloc[: ,[0]]


#When using [], the result is a Pandas DataFrame.
```

***Why is this a Series?***
- When you pass a single integer (not a list) as the column index, Pandas returns a Series (a one-dimensional array-like object).

## **2.4. Data Normalization**

Normalization scales the numerical data so that all the values are between 0 and 1. This is important for machine learning tasks where features need to be on the same scale.


Task: Create data frame from URL and applying column normalization

*Each value in the columns is normalized by subtracting the column's minimum value and dividing by the range (max - min).*

image.png

```python
import pandas as pd
mydata = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
mydata.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

mydataNum = mydata.iloc[:,0:4]

mydata_normalized = (mydataNum - mydataNum.min()) / (mydataNum.max()-mydataNum.min())
mydata_normalized
```

```python
grades_dict={
             'Name':["A","C","V","H","T","U","F","L","Z","B"],
             'Grade':[8,9,10,7,4,5,6,9,8,8]
             }
#OUTPUT: Mean grade is  7.4  and median grade is  8.0
```

```python
#Task-1: Given a dataset of student grades, we calculate the mean and median.

grades_dict={'Name':["A","C","V","H","T","U","F","L","Z","B"], 'Grade':[8,9,10,7,4,5,6,9,8,8]}
gradesFrame = pd.DataFrame(grades_dict)
gradesFrame

meanGrade = gradesFrame['Grade'].mean()
medianGrade = gradesFrame['Grade'].median()

print("Mean grade is ",meanGrade,' and median grade is ',medianGrade)
```

```python
# Task-3: Filter the students above average
grades_dict={'Name':["A","C","V","H","T","U","F","L","Z","B"], 'Grade':[8,9,10,7,4,5,6,9,8,8]}
gradesFrame = pd.DataFrame(grades_dict)
print(gradesFrame)

meanGrade = gradesFrame['Grade'].mean()

topStudentsFrame = gradesFrame[gradesFrame['Grade']> meanGrade]
topStudentsFrame
```
